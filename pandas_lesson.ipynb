## Accessing Data

```python
# To slice a pandas dataframe using row and column numbers
# detaframe_name.iloc[row_slice, column_slice]
periodic_data.iloc[35]
```

```python
periodic_data.iloc[35,:3]
```

```python
# USe the iloc function to
# select the 5th row
# select row 30 to the end and the first three columns
# select row 50 to the end every other row and columns 2 through 4
```

```python
periodic_data.iloc[30:,:3]
```

```python
# syntax for intervals {start:stop:interval}
periodic_data.iloc[50::2,2:5]
```
